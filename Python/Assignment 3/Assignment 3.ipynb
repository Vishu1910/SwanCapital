```python
import pandas as pd
inputt=r'C:\Users\admin\Desktop\Dmart.csv'
df = pd.read_csv(inputt)
```

```python
df['ROC']=((df['Close']-df['Open'])/df['Open'])*100
```

```python
df.loc[df['ROC']<-0.5,"Type of Trade"]='0'
df.loc[df['ROC']>0.5,"Type of Trade"]='1'
```

```python
Winners = pd.DataFrame(columns=["Ticker","Date/Time","Open","Low","High","Close","ROC"])
Losers = pd.DataFrame(columns=["Ticker","Date/Time","Open","Low","High","Close","ROC"])
```

```python
Winners = pd.DataFrame()
Losers = pd.DataFrame()
k=0
while(k<len(df['ROC'])):
    for i,j in df.iterrows():
        if(df.loc[i,'Type of Trade'] == '1'):
            Winners = Winners.append(df.loc[i,["Ticker","Date/Time","Open","Low","High","Close","ROC"]])
        k+=1
        if(df.loc[i,'Type of Trade']== '0'):
            Losers = Losers.append(df.loc[i,["Ticker","Date/Time","Open","Low","High","Close","ROC"]])
        k+=1    
            
```

```python
display(Winners)
display(Losers)
```

```python
print("Total number of trades where ROC > 0.5 = ",len(Winners))
print("Total number of trades where ROC < -0.5 = ",len(Losers))
```
