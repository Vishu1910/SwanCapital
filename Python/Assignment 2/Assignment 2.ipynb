Using the Dmart csv which contains the OHLC data for Dmart for last 1 year, perform the following things in python:

1. Create a new empty Dataframe 'Winners'  with Columns [Date, Open, Close, ROC].
2. Create a new empty Dataframe 'Losers'  with Columns [Date, Open, Close, ROC].
3. Using for loop, append only the winning trades in 'Winners' Dataframe and only the Losing Trades in 'Losers' Dataframe
4. Calculate the total ROC of 'Winners' and 'Losers' separately
5. Print Step 3 and 4  

```python
import pandas as pd
inputt = r'C:\Users\admin\Desktop\Dmart.csv'
df = pd.read_csv(inputt)
```

```python
df['ROC']=((df['Close']-df['Open'])/df['Open'])*100
```

```python
x=df.loc[df['ROC']<=0,'Type Of Trade']='0'
y=df.loc[df['ROC']>0,'Type Of Trade']='1'
```

```python
Winners = pd.DataFrame(columns=["Date/Time","Open","Close","ROC"])
Losers = pd.DataFrame(columns=["Date/Time","Open","Close","ROC"])
```

```python
Winners = pd.DataFrame(columns=["Date/Time","Open","Close","ROC"])
Losers = pd.DataFrame(columns=["Date/Time","Open","Close","ROC"])

Winners = pd.DataFrame()
Losers = pd.DataFrame()
for i,j in df.iterrows():
#     print(df.loc[i, 'Type Of Trade']=='1')
    
    if(df.loc[i, 'Type Of Trade'] == '1'):
        Winners = Winners.append(df.loc[i,['Date/Time', 'Open', 'Close', 'ROC']])
    if(df.loc[i, 'Type Of Trade'] == '0'):
        Losers = Losers.append(df.loc[i,['Date/Time', 'Open', 'Close', 'ROC']])    
```

```python
display(Winners)
display(Losers)
```

```python
print("Total ROC of winners combined is",Winners['ROC'].sum(axis=0))
print("Total ROC of losers combined is",Losers['ROC'].sum(axis=0))
```
