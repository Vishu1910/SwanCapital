```python
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
import pandas as pd
inputt=r'C:\Users\admin\Desktop\Dmart.csv'
df = pd.read_csv(inputt)
```

```python
df['ROC']=((df['Close']-df['Open'])/df['Open'])*100
```

```python
df.loc[df['ROC']<-0.5,"Type of Trade"]='0'
df.loc[df['ROC']>0.5,"Type of Trade"]='1'
```

```python
Winners = df[df['ROC']>0.5]
Losers = df[df['ROC']<-0.5]
display(Winners)
display(Losers)
```

```python
print("Total number of trades where ROC < -0.5 ",len(Losers))
print("Total number of trades where ROC > 0.5 ",len(Winners))
```

```python
from datetime import datetime
df['Month']=pd.to_datetime(df['Date/Time']).dt.month
print(df.groupby(['Month'])['ROC'].sum())
```
