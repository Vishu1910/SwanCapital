```python
import pandas as pd
import numpy as np
inputt=r'C:\Users\admin\Desktop\Dmart.csv'
df = pd.read_csv(inputt)
```

```python
from datetime import datetime
df['Month']=pd.to_datetime(df['Date/Time']).dt.month
df['Date/Time']=pd.to_datetime(df['Date/Time']).dt.date
#display(df)
```

```python
df['Week']=pd.to_datetime(df['Date/Time']).dt.week
display(df)
df.to_csv(r'C:\Users\admin\Desktop\df.csv',index=False)
```

```python
df_1 = pd.DataFrame()
df_2 = pd.DataFrame()
df_1 = df.groupby(['Week','Month'])['Open'].nth(0)
df_2 = df.groupby(['Week','Month'])['Close'].nth(-1)
df_filter = pd.concat([df_1,df_2],axis=1)
#display(df.groupby(['Week','Month'])['Open'].nth(0))
df_filter['ROC']=((df_filter['Close']/df_filter['Open'])-1)*100
display(df_filter)
display(df_filter['ROC'].sum())
```

```python
table = pd.pivot_table(df_filter,values='ROC',index=['Month','Week'],columns='Month',aggfunc=np.sum)
display(table)
```
