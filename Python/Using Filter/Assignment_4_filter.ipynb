```python
import pandas as pd
import numpy as np
inputt=r'C:\Users\admin\Desktop\Dmart.csv'
df = pd.read_csv(inputt)
```

```python
from datetime import datetime
df['Month']=pd.to_datetime(df['Date/Time']).dt.month
df['Date/Time']=pd.to_datetime(df['Date/Time']).dt.date
#display(df)
```

```python
import datetime
df['day_of_month'] = pd.to_datetime(df['Date/Time']).dt.day
#df['week_number'] = (df['day_of_month'] - 1) // 7 + 1
#display(df)
df['Week']=pd.to_datetime(df['Date/Time']).dt.week
display(df)
df.to_csv(r'C:\Users\admin\Desktop\test.csv')
```

```python
week=1
df.loc[0,'NewWeek']=1
for i in range(1,len(df)):
    if(df.loc[i,'Month']!=df.loc[i-1,'Month']):
        week=1
        df.loc[i,'NewWeek']=week
    elif(df.loc[i,'Month']==df.loc[i-1,'Month']):
        if(df.loc[i,'Week']!=df.loc[i-1,'Week']):
            week+=1
            df.loc[i,'NewWeek']=week
        else:
            df.loc[i,'NewWeek']=df.loc[i-1,'NewWeek']
display(df)
df.to_csv(r'C:\Users\admin\Desktop\trial.csv')
```

```python
df_1 = pd.DataFrame()
df_2 = pd.DataFrame()
df_1 = df.groupby(['Month','NewWeek'])['Open'].nth(0)
df_2 = df.groupby(['Month','NewWeek'])['Close'].nth(-1)
df_filter = pd.concat([df_1,df_2],axis=1)
df_filter['ROC']=((df_filter['Close']/df_filter['Open'])-1)*100
display(df_filter)
display(df_filter['ROC'].sum())
```

```python
table = pd.pivot_table(df_filter,values='ROC',index='NewWeek',columns='Month',dropna=True)
display(table)
```
