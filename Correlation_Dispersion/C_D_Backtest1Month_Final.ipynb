```python
tradesheet = pd.DataFrame()
tradesheet = stock
tradesheet.loc[100,'EQUITYCLOSE']=16170.15
display(tradesheet)
```

```python
#display(stock)
```

```python
for i in range(len(tradesheet)):
    tradesheet.loc[i,'ContractValue']=tradesheet.loc[i,'EQUITYCLOSE']*int(tradesheet.loc[i,'Lot'])
    tradesheet.loc[i,'Margin']=0.20*(tradesheet.loc[i,'ContractValue'])
    if(tradesheet.loc[i,'INSTRUMENT']=='OPTSTK'):
        tradesheet.loc[i,'NumberOfLots']=round(700000/tradesheet.loc[i,'Margin'])
        tradesheet.loc[i,'Signal']='Short'
    else:
        tradesheet.loc[i,'NumberOfLots']=round(30000000/tradesheet.loc[i,'Margin'])
        tradesheet.loc[i,'Signal']='Long'
```

```python
for i in range(len(tradesheet)):
    tradesheet.loc[i,'MarginPrice']=0.20*float(tradesheet.loc[i,'EQUITYCLOSE'])*int(tradesheet.loc[i,'Lot'])*int(tradesheet.loc[i,'NumberOfLots'])
    tradesheet.loc[i,'EntryPrice']=float(tradesheet.loc[i,'CLOSE'])*int(tradesheet.loc[i,'Lot'])*int(tradesheet.loc[i,'NumberOfLots'])
    tradesheet.loc[i,'ExitPriceee']=float(tradesheet.loc[i,'EXITPRICE'])*int(tradesheet.loc[i,'Lot'])*int(tradesheet.loc[i,'NumberOfLots'])
    
    if(tradesheet.loc[i,'INSTRUMENT']=='OPTSTK'):
        tradesheet.loc[i,'Profit/Loss']=tradesheet.loc[i,'EntryPrice']-tradesheet.loc[i,'ExitPriceee']
    else:
        tradesheet.loc[i,'Profit/Loss']=tradesheet.loc[i,'ExitPriceee']-tradesheet.loc[i,'EntryPrice']
display(tradesheet)
```
